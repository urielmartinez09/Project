# Where The Data Comes From
 Our data is taken from Kaggle which the data was compiled through an anonymous survey taken from January to June of 2023. The data was taken in India. The surveyees were between the ages of 18-60 and volunteered from different cities, no professional mental health assessments were taken before the survey. The point of the survey was originally to provide insights into how everyday factors might correlate with mental health risks. There are 20 variables and 140,700 observations.

```python
# Sebastians Graph
DF_Depression_Influence = (
DF.value_counts(subset=['Study Satisfaction', 'Depression']).reset_index(name='count').sort_values(by='Study Satisfaction', ascending=True)

)

fig = px.pie(DF_Depression_Influence, 
    values='count', 
    names = 'Depression',
    facet_col='Study Satisfaction',
    color = 'Depression',
    facet_col_wrap = 3,
)


fig.update_layout(
                  title='How Do Different Factors Influence Depression Rates',
                  template='ggplot2',
                  legend_title='Depression',
                  width=1000,
                  height=700,
                      
                      )

fig.show()
```
